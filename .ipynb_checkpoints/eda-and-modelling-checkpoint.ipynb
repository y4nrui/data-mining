# EDA

```python
# check for missing values
data.isnull().any()
```

```python
# unique values of town column
data['town'].value_counts()
```

```python
data['flat_type'].value_counts()
```

```python
data['flat_model'].value_counts()
```

```python
# histogram of resale_price
sns.distplot(data['resale_price'])
```

```python
# check for multicollinearity
# resale_price and floor_area_sqm are related
sns.pairplot(data, hue='town', diag_kind='hist')
```

```python
# timeseries of resale_price
plt.plot(data['month'], data['resale_price'])
plt.figure(figsize=(100,100))
plt.show()
```

```python
fig = sns.boxplot(x='month', y='resale_price', data=data)
```

```python
data['town'].value_counts()
```

```python
# box plot of town to resale_price
plt.figure(figsize=(50,15))
fig = sns.boxplot(x='town', y='resale_price', data=data)
```

```python
# box plot of flat_model to resale prices
plt.figure(figsize=(50,15))
fig = sns.boxplot(x='flat_model', y='resale_price', data=data)
```
